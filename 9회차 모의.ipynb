2. 각 연도별 최대 검거율을 가진 범죄유형을 찾아서 해당 연도 및 유형의 검거건수들의 총합을 구하시오.  
(검거율 = 검거건수 / 발생건수) 

```python
import pandas as pd 

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_1_2.csv")
display(df.head(2))

crime_cols = [f'범죄유형{i}' for i in range(1,11)]
df2 = df.groupby(['연도','구분'])[crime_cols].sum().reset_index()
display(df2.head(5))

df3 = df[df['구분']=='검거건수'].set_index('연도').drop(columns='구분')
df4 = df[df['구분']=='발생건수'].set_index('연도').drop(columns='구분')
dfratio = df3/df4

df_max = dfratio.idxmax(axis=1) # 각 연도별 최대 검거율 가진 범죄유형
dfmax_reset = df_max.reset_index().rename({0: '범죄유형'},axis=1)
display(dfmax_reset)
```

3. 제시된 문제를 순서대로 풀고, 해답을 제시하시오 

```python
import pandas as pd 
import numpy as np
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_1_3.csv")
#display(df.head(3))
#print(df.isnull().sum())

# 평균만족도 : 결측치는 평균만족도 컬럼의 전체 평균으로 채우시오
satify_mean = df['평균만족도'].mean()
df['평균만족도'].fillna(satify_mean)

# 근속연수 : 결측치는 각 부서와 등급별 평균 근속연수로 채우시오. (평균값의 소숫점은 버림 처리)
df2 = df.groupby(['부서','등급'])['근속연수'].mean().reset_index()
df = df.merge(df2,on=['부서','등급'],how='left')
display(df)

import pandas as pd
import numpy as np

mean_tenure=(
    df.groupby(["부서","등급"])["근속연수"]
    .mean()
    .apply(np.floor)
    .reset_index()
    .rename(columns={"근속연수":"평균근속연수"})
)
df = df.merge(mean_tenure,on=["부서","등급"],how="left")
df["근속연수"]=np.where(
    df["근속연수"].isnull(),
    df["평균근속연수"],
    df["근속연수"]
)
df.drop(columns=["평균근속연수"],inplace=True)
display(df)
```

# 작업형 제2유형 

```python
# 훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오. (Target : 농업 유형 (다중분류, {0, 1, 2}))
# 제출형식은 ID, pred 두 칼럼만 존재해야한다. (평가 지표 : Macro F1 Score)
# 9_2_train, 9_2_test 

import pandas as pd 
import numpy as np
train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_2_test.csv")

```

```python
# EDA
display(train.head(3))
display(test.head(3))
print("\n")
print("Train 데이터 null 여부")
print(train.isnull().sum(),"\n")
print("Train 데이터 Information")
print(train.info())
```

```python
from sklearn.model_selection import train_test_split

test_id = test['ID']
train_X = train.drop(['라벨'],axis=1)
train_y = train['라벨']

test_X = test.drop(['라벨'],axis=1)
test_y = test['라벨']

train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y,test_size=0.3, random_state=42)

# 데이터 전처리
cat_columns = train_X.select_dtypes('object').columns.to_list()
```

```python
# 원-핫 인코딩
from sklearn.preprocessing import OneHotEncoder

Onehotencoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

train_X_preprocessed = Onehotencoder.fit_transform(train_X[cat_columns])
valid_X_preprocessed = Onehotencoder.transform(valid_X[cat_columns])
test_X_preprocessed = Onehotencoder.transform(test_X[cat_columns])
```

```python
# 모델 적합
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(train_X_preprocessed,train_y)

# 모형 성능 확인
from sklearn.metrics import f1_score 
pred_val = rf.predict(valid_X_preprocessed)
print("valid f1-macro : ",f1_score(valid_y, pred_val, average='macro'))
```

```python
# 최종 예측
test_pred = rf.predict(test_X_preprocessed)
test_pred = pd.DataFrame(test_pred, columns=['pred'])
```

```python
result = pd.concat([test_id, test_pred],axis=1)
print(result.head(2))
```

```python
# 테스트 데이터로 예측
result = pd.concat([test_id,test_pred],axis=1)
#result.to_csv('9_2_result.csv',index=False)
display(result)
```

# 작업형 제3유형

한 제조 회사에서 생산성을 높이고자 직원들의 주요 생산성 요인을 분석하기로 결정하였다. 이를 위해 200명의 직원 데이터를 수집했으며, 직원들의 근무 기간, 특성 정보, 그리고 개인적인 속성을 조사하였다.  

(1) 모든 칼럼을 활용하여 design 을 예측하는 다중회귀 분석을 시행한 후 유의하지 않은 설명변수 개수를 구하시오. (패널티는 포함되지 않으며, 모델의 절편항은 포함)

(2) 훈련데이터 (학습용 데이터) 의 예측값과 실제값의 피어슨 상관계수를 구하시오 
(훈련데이터 : id1 ~ 140, 테스트 데이터 : id41 ~ 200) 


(3) 적합한 모델을 활용하여 테스트 데이터에서의 RMSE 를 구하시오

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_3_1.csv")
display(df.head(3))

train = df[:40]
test = df[40:]
```

```python
from statsmodels.formula.api import ols 

formula = "design ~ tenure + f2 + f3 + f4 + f5"
model = ols(formula,data=train).fit()

p_value = model.pvalues 

non_significant = (p_value[1:]>=0.05).sum()
print("유의하지 않은 설명변수의 갯수 : ",non_significant)

y_pred = model.predict(train)
y_real = train['design']

from scipy.stats import pearsonr 
correlation, _ = pearsonr(y_pred,y_real)
print("상관계수 : ",correlation)

y_pred_test = model.predict(test)
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(test['design'],y_pred_test)
print("RMSE : ",rmse)
```

한 통신회사에서는 고객 이탈을 줄이고자 주요 요인들을 분석하기로 결정하였다. 이를 위해 500 명의 고객 데이터를 수집했으며, 고객의 서비스 이용 및 가입 정보, 그리고 일부 개인적인 속성을 조사하였다.  

(1) 고객 이탈을 예측하는 로지스틱 회귀를 시행한 후 col1 칼럼의 p-value 를 구하시오

(2) 폰 서비스를 받지 않은 고객 대비 받은 고객의 이탈 확률 오즈비를 구하시오  

(3) 이탈할 확률이 0.3 이상인 고객 수를 구하시오  

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_3_2.csv")
display(df.head(3))
```

```python
from statsmodels.formula.api import logit

formula = "churn ~ col2 + Phone_Service + Tech_Insurance + col1"
model = logit(formula,data=df).fit()

print("col1 의 p-value : ",round(model.pvalues['col1'],3))

oldds = round(np.exp(model.params['Phone_Service']),3)
print("오즈비 : ",oldds)

# 이탈할 확률이 0.3 이상 
model_prob = model.predict(df)
prob03 = sum(model_prob>0.3)
print("이탈할 확률이 0.3 이상인 고객수 : ",prob03)
```
