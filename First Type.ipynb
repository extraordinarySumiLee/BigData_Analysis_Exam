# 함수 모음

```python
x = 3.1415
y = x * 3
z = [1, 2, 3, 4, 5]

print("x의 절댓값 : ", abs(x)) 
print("x의 소숫점 3째자리까지 : ", round(x, 3)) 
print("x, y 중 최댓값 : ", max(x, y))
print("x, y 중 최솟값 : ", min(x, y)) 
print("z의 합 : ", sum(z))
print("x^y : ", pow(x, y))
print("y를 x로 나눈 몫과 나머지 : ", divmod(y, x)) # y를 x로 나눈 몫과 나머지

ab = [True, True, False]
print("\n--- 내장함수 ---")
print("ab의 모든 요소가 참인지 검사 : ", all(ab))
print("ab의 하나 이상의 요소가 참인지 검사 : ", any(ab))

# 💡 객체 반환 함수들은 list()로 감싸서 출력하도록 수정
print("z의 모든 요소를 인덱스와 함께 반환 : ", list(enumerate(z, start=0)))
print("function의 조건을 만족하는 요소만 필터링 : ", list(filter(lambda v: v < 5, z)))
print("function을 요소 전체에 적용하여 변환 : ", list(map(lambda v: v * 2, z)))

# 💡 zip은 두 개 이상의 iterable을 묶을 때 사용하므로 변수 추가 및 list() 적용
z_letters = ['A', 'B', 'C', 'D', 'E']
print("여러 iterable의 요소를 병렬로 묶어서 반환 : ", list(zip(z, z_letters)))
```
