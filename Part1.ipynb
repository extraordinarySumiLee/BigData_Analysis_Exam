# Section 02 데이터 타입의 이해

```python
# 집합(Set) 형 데이터 타입 : 순서가 없고 중복된 요소를 허용하지 않는 컬렉션
# 수학적인 집합 연산을 지원하며, 주로 중복을 제거하거나 항목의 존재 여부를 빠르게 검사할 때 사용된다.

fruits = {'apple','banana','cherry','apple'}
print("Fruits set : ",fruits)

# 빈 집합 생성
empty_set = set()
print("Empty set : ",empty_set)
```

```python
# 요소 추가
fruits.add('orange')
print("After adding orange:",fruits)

# 요소 삭제
fruits.remove('banana')
print("After deleting banana : ",fruits)

# 집합간 연산
other_fruits = {'berry','watermelon'}
union_fruits = fruits.union(other_fruits)
intersection_fruits = fruits.intersection(other_fruits)
print("Union of fruits:",union_fruits)
print("Intersection of fruits:",intersection_fruits)
```

자료형별 메서드 요약
- str
    - upper(), lower(), replace(old, new), split(sep=None), join(iterable), strip(), istrip(), rstrip()
    - find(sub) : 부분 문자열 sub 를 찾아 처음 발견된 인덱스를 반환. 찾지 못하면 -1 을 반환
    - count(sub) : 부분 문자열 sub 가 문자열 내에 몇 번 등장하는지 횟수를 반환 
- list
    - append(item), insert(index, item), extend(iterable)
    - remove(item) : 리스트에서 첫 번째로 발견된 item 을 제거. 없으면 ValueError 발생
    - pop(index=-1) : 지정한 인덱스 위치의 요소를 제거하고, 그 요소를 반환. 기본값 (-1) 은 마지막 요소
    - clear()
    - index(item) : item 이 처음으로 나타내는 인덱스를 반환. 없으면 ValueError 발생 
    - count(item), sort(key=None, reverse=False), reverse(), copy()
- tuple
    - count(item), index(item)
- dictionary 
    - get(key[,defualt]) : key 에 대응하는 값 반환. key 가 없으면 default 를 반환하고, 기본값이 없으면 None 반환 
    - keys(), values()
    - items() : 키, 값 쌍 반환
    - update([other]) : 다른 딕셔너리(other) 의 키-값 쌍을 현재 딕셔너리에 병합
    - pop(key[, default])
    - popitem() : 마지막으로 추가된 (키, 값) 쌍을 꺼내고, 딕셔너리에서 삭제
    - setdefault(key[,defualt]) : key 가 이미 있으면 해당 값을 반환, 없으면 default 값을 키와 함께 딕셔너리에 추가한 후 그 값을 반환
    - clear()

```python
a = np.array([21, 31, 58])
b = np.array([24, 44, 67])
c = np.empty(a.size+b.size,dtype=a.dtype)

c[0::2] = a
c[1::2] = b 
print(c)
```
