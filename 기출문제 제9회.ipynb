### 작업형 제 1유형

1. 각 연도 및 성별의 총 대출액의 절댓값 차이를 구하고, 절댓값 차이가 가장 큰 지역코드를 구하시오

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

raw_data = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_1_1.csv")
display(raw_data)

pd.set_option('display.max_columns',None) # 모든 칼럼이 출력되게 조절

raw_data['총대출액'] = raw_data['금액1'] + raw_data['금액2']
# 각 연도 및 성별로 총 대출액 합계 계산
dfgender = raw_data.groupby(['year','gender','지역코드'])['총대출액'].sum().reset_index()
display(dfgender)
# reset_index : 인덱스로 가버린 연도, 성별, 지역코드를 다시 일반 컬럼으로 끄집어내기
```

```python
# 각 연도 및 성별에 따른 차이를 계산하기 위해 피벗 테이블 생성
df_pivot = dfgender.pivot_table(index= ['year','지역코드'], columns='gender',values='총대출액',fill_value=0)
display(df_pivot)
# index : 표의 맨 왼쪽 '행' 자리에 올 기준 정함 
# columns : 표의 열 자리에 펼쳐질 기준을 정함
# values : 행과 열이 만나는 교차에 채워넣을 진짜 데이터 값
# fill_value = 0 : 만약 특정 연도/ 지역에 특정 성별의 데이터가 없어서 결측치가 생기면 0으로 채워라.

# groupby 는 결과를 위에서 아래로 길게 보여주고, pivot_table은 데이터를 가로와 세로 양방향으로 쫙 펼쳐줌
```

```python
# 성별에 따른 절댓값 차이를 계산하고, 차이가 가장 큰 지역 코드를 출력
df_pivot['abs_diff'] = abs(df_pivot[0]-df_pivot[1])

max_diff = df_pivot['abs_diff'].idxmax()

max_diff_region = max_diff[1]

print("차이가 가장 큰 지역코드 : ",max_diff_region)
display(df_pivot)
```

2. 각 연도별 최대 검거율을 가진 범죄유형을 찾아서 해당 연도 및 유형의 검거건수들의 총합을 구하시오  
(검거율 = 검거건수/발생건수)

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_1_2.csv")
display(df)
```

```python
df1 = df[df["구분"]=="검거건수"].set_index("연도").drop(columns="구분")
df2 = df[df["구분"]=="발생건수"].set_index("연도").drop(columns="구분")

# 검거율 계산
dfratio = df1/df2
display(dfratio)
```

```python
# 각 연도별 최대 검거율을 가진 범죄유형 산출
dfmax = dfratio.idxmax(axis=1)
dfmax_reset = dfmax.reset_index().rename({0: '범죄유형'},axis=1)
display(dfmax_reset)
```

```python
# 각 연도별 범죄유형에 해당하는 검거건수의 총합을 출력
df1_melt = df1.reset_index().melt(id_vars="연도",var_name="범죄유형",value_name="검거건수")
display(df1_melt)
# melt : 가로로 뚱뚱하게 옆으로 펼져져 있는 표를 녹여서 아래로 길게 늘어뜨리는 역할
# id_vars = : 식별자로 사용할 컬럼. 녹이지 않고 그대로 고정해 둘 기준 컬럼
# var_name : 가로로 펼쳐져 있던 컬럼 이름들이 세로로 합쳐지며 들어갈 새로운 컬럼의 이름 
# value_name : 표 안에 진짜 숫자 데이터들이 모딜 새로운 컬럼의 이름 
```

```python
dffinal = pd.merge(dfmax_reset, df1_melt, on = ["연도","범죄유형"], how='left')
display(dffinal)
# pd.merge : 두 개의 서로 다른 데이터프레임을 특정 기준을 가지고 하나로 합치는 힘수. 엑셀의 VLOOKUP 이나 SQL 의 JOIN 역할 
# dfmax_reset : 병합의 기준이 되는 데이터프레임
# df1_melt : 붙이고 싶은 추가 정보를 가지고 있는 데이터프레임
# on : 두 표를 합칠 때 사용할 공통된 연결고리. 
# how : 왼쪽 데이터를 주인공으로 삼겠다는 뜻
```

```python
total_sum = dffinal['검거건수'].sum()
print("총 검거건수 합계 =",total_sum)
```

3. 제시된 문제를 순서대로 풀고, 해답을 제시하시오

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_1_3.csv")
```

결측치 처리 : 결측치는 평균만족도 컬럼의 전체 평균으로 채우시오  
근속연수 : 결측치는 각 부서와 등급별 평균 근속연수로 채우시오. (평균값의 소수점은 버림 처리)  

```python
display(df)
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 평균만족도 칼럼은 해당 칼럼의 평균으로 대치
df["평균만족도"] = df["평균만족도"].fillna(df["평균만족도"].mean())
```

```python
# 부서와 등급에 따라 평균 근속연수로 대치
mean_tenure = (
    df.groupby(["부서","등급"])["근속연수"]
    .mean()
    .apply(np.floor)
    .reset_index()
    .rename(columns={"근속연수":"평균근속연수"})
)
```

```python
# merge 로 결측치 대치
df = df.merge(mean_tenure, on = ["부서","등급"],how="left")
df["근속연수"] = np.where(
    df["근속연수"].isnull(), # 조건 : 근속연수가 결측치인 경우
    df["평균근속연수"], # 결측치일때 대치값
    df["근속연수"] # 결측치가 아닐 때 기존 값 유지
)

df.drop(columns=["평균근속연수"],inplace=True)
```

```python
display(df)
```

(1) 부서가 'HR' 이고 등급이 'A' 인 사람들의 평균 근속연수를 계산하시오

```python
A = df.loc[(df["부서"]=="HR")&(df["등급"]=="A"),"근속연수"].mean()
print(A)
```

(2) 부서가 Sales 이고 등급이 B 인 사람들의 평균 교육 참가 횟수

```python
B = df.loc[(df["부서"]=="Sales")&(df["등급"]=="B"),"교육참가횟수"].mean()
print(B)
```

(3) 두 조건의 합

```python
print(A+B)
```

### 작업형 제 2 유형

훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오  
(Target : 농업 유형 (다중분류, {0,1,2}))  
제출 형식은 ID, pred 두 칼럼만 존재해야한다.

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_2_train.csv")
df2 = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/9_2_test.csv")
display(df)
```

```python
import pandas as pd 
import numpy as np

train = df 
test = df2 
```

```python
# 데이터 탐색 
print(train.info())
```

```python
print(test.info())
```

```python
# 테스트 데이터를 활용한 최종 결과를 제출하기 위해 ID 칼럼을 따로 저장한다. 
test_id = test['ID']
train_X = train.drop(['라벨'],axis=1)
train_y = train['라벨']

test_X = test.drop(['라벨'],axis=1)
test_y = test['라벨']
```

```python
# 데이터 분할 
from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y, test_size=0.3, random_state=1)
```

```python
# 데이터 전처리
cat_columns = train_X.select_dtypes('object').columns.to_list()
# num_columns = train_X.select_dtypes('number').columns.to_list()
```

```python
# one-hot incoding
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

train_X_preprocessed = onehotencoder.fit_transform(train_X[cat_columns])
valid_X_preprocessed = onehotencoder.transform(valid_X[cat_columns])
test_X_preprocessed = onehotencoder.transform(test_X[cat_columns])

# 모델 적합
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1)
rf.fit(train_X_preprocessed,train_y)

# 모형 성능 확인
from sklearn.metrics import f1_score
pred_val = rf.predict(valid_X_preprocessed)
print('valid f1-macro:',f1_score(valid_y, pred_val, average='macro'))
```

```python
# 테스트 데이터 활용한 최종 예측
test_pred = rf.predict(test_X_preprocessed)
test_pred = pd.DataFrame(test_pred, columns=['pred'])
```

```python
# ID, pred 칼럼만 존재하는 결과 파일 생성
result = pd.concat([test_id, test_pred], axis=1)
print(result.head(2))
```

```python
# 최종 결과 저장
result.to_csv('9_2_result.csv', index=False) # 시험에서 제시되는 파일명 작성할 것
```

```python
# 모델 성능이 낮으면 교차 검증을 활용한 하이퍼파라미터 튜닝을 진행 
train_X_full = np.concatenate([train_X_preprocessed, valid_X_preprocessed],axis=0)
train_y_full = np.concatenate([train_y, valid_y], axis=0)
```

```python
# GridSearchCGV 를 통해 하이퍼파라미터 튜닝 진행
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': [10, 20, 30],
              'min_samples_split' : [2, 5, 10]}

rf = RandomForestClassifier(random_state=1)
rf_search = GridSearchCV(estimator = rf,
                         param_grid = param_grid, 
                         cv = 3,
                         scoring = 'f1_macro')

rf_search.fit(train_X_full, train_y_full)
GridSearchCV(cv=3, estimator = RandomForestClassifier(random_state=1),
             param_grid={'max_depth': [10, 20, 30],
                         'min_samples_split' : [2, 5, 10]},
                         scoring = 'f1_macro')

print("교차검증 MAE-score:", rf_search.best_score_)
```

```python
# 하이퍼파라미터 튜닝 결과를 바탕으로 테스트 데이터 활용하여 최종 예측 
test_pred2 = rf_search.predict(test_X_preprocessed)
test_pred2 = pd.DataFrame(test_pred2, columns=['pred'])
```

```python
# 테스트 데이터로 예측 
result = pd.concat([test_id, test_pred2],axis=1)

# 최종 결과 저장
result.to_csv('9_2_result.csv', index=False) 
```
