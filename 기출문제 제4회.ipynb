# 작업형 제2유형

훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오. (Target : 이탈 여부 (Attrition_Flag) 가 Attrited Customer 일 확률)  
* 제출 형식은 ID, prob 두 칼럼만 존재해야 한다. (평가지표 : AUC)


```python
import pandas as pd 
train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_2_test.csv")
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
display(train.head(3))
print(train.info())
```

```python
import pandas as pd 
train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_2_test.csv")
#display(train.head(3))

# target 은 자체 onehot
print(train['Attrition_Flag'].unique())
def target(df) : 
    if df == "Existing Customer" :
        return 0
    elif df == "Attrited Customer" :
        return 1
train['Attrition_Flag'] = train['Attrition_Flag'].apply(target)
test['Attrition_Flag'] = test['Attrition_Flag'].apply(target)

train_X = train.drop(['ID'],axis=1)
train_y = train['Attrition_Flag']
test_X = test.drop(['ID'],axis=1)
test_y = test['Attrition_Flag']

from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_cols = train_X.select_dtypes('object').columns.to_list()
num_cols = train_X.select_dtypes('number').columns.to_list()

train_X_onehot = onehot.fit_transform(train[cat_cols])
test_X_onehot = onehot.transform(test[cat_cols])

import numpy as np 
train_X_pre = np.concatenate([train_X_onehot,train_X[num_cols]],axis=1)
test_X_pre = np.concatenate([test_X_onehot,test_X[num_cols]],axis=1)

# RandomForest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(train_X_pre,train_y)
prediction = rf.predict_proba(test_X_pre)[:,1]

from sklearn.metrics import roc_auc_score
auc = roc_auc_score(test_y,prediction)

print(f"AUC : {round(auc,1)}")

test_id = test['ID'] 
result = pd.DataFrame({'ID':test_id,'proba':prediction})

display(result.head(5))
```

# 작업형 제3유형  


어느 회사에서 두 부서 (A와 B) 직원들의 주간 근무 시간 평균이 유믜미하게 다른지를 조사하고자 한다. 각 부서에서 각각 30명의 직원을 랜덤으로 추출하여 주간 근무 시간을 조사하였다. 

(1) 두 부서의 평균과 표춘편차를 구하시오. (소수점 둘째 자리까지 반올림)  
(2) 두 부서의 평균 근무시간이 유의미하게 다른지 검정하기 위해 독립 표본 t-검정을 수행하고, 검정통계량을 계산하시오. (소수점 둘째 자리까지 반올림)  
(3) p-value 를 바탕으로 유의수준 5% 에서 귀무가설의 기각/채택 여부를 결정하시오. (p-value 는 소수점 둘째 자리까지 반올림)

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_3_1.csv")
display(df.head(3))
```

```python
print("A 부서의 주간 근무시간")
df_A = df[(df['department']=="A")]
df_B = df[(df['department']=="B")]
A_mean = df_A['hours_worked'].mean()
A_std = df_A['hours_worked'].std()
B_mean = df_B['hours_worked'].mean()
B_std = df_B['hours_worked'].std()
print(f"평균 : {round(A_mean,2)}, 표준편차 : {round(A_std,2)}")
print("B 부서의 주간 근무시간")
print(f"평균 : {round(B_mean,2)}, 표준편차 : {round(B_std,2)}")
```

```python
from scipy.stats import ttest_ind 
# 독립표본 t-검정
statistic, pvalue = ttest_ind(df_A['hours_worked'],df_B['hours_worked'])
print(f"검정통계량 : {round(statistic,2)}")
if pvalue < 0.05 : 
    print("귀무가설을 기각합니다.")
else :
    print("귀무가설을 채택합니다.")
```

어느 제조업체에서 세 공장 (A, B, C) 에서 생산된 제품의 품질 점수가 유의미하게 다른지를 조사하고자 한다. 각 공장에서 각각 30개의 제품을 랜덤으로 추출하여 품질 점수를 조사하였다.  

(1) 세 공장 품질 점수의 평균과 표준편차를 구하시오. (소수점 둘째 자리까지 반올림)    
(2) 두 부서의 평균 근무 시간이 유의미하게 다른지 검정하기 위해 크루스칼-왈리스 검정을 수행하고, 검정통계량을 계산하시오. (소수점 둘째 자리까지 반올림)  
(3) p-value 를 바탕으로 유의수준 5% 에서 귀무가설의 기각/채택 여부를 결정하시오. (p-value 는 둘째자리까지 반올림)

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/4_3_2.csv")
display(df.head(3))
```

```python
df_A = df[(df['factory']=='A')]
df_B = df[(df['factory']=='B')]
df_C = df[(df['factory']=='C')]

print(f"공장 A의 표준 : {round(df_A['quality_score'].mean(),2)}, 표준편차 : {round(df_A['quality_score'].std(),2)}")
print(f"공장 B의 표준 : {round(df_B['quality_score'].mean(),2)}, 표준편차 : {round(df_B['quality_score'].std(),2)}")
print(f"공장 C의 표준 : {round(df_C['quality_score'].mean(),2)}, 표준편차 : {round(df_C['quality_score'].std(),2)}")

# 크루스칼-왈리스검정
group1,group2,group3 = df_A['quality_score'],df_B['quality_score'],df_C['quality_score']
from scipy.stats import kruskal
statistic, pvalue = kruskal(group1,group2,group3)
print(f"검정통계량 : {round(statistic,2)}")
if pvalue < 0.05 : 
    print("귀무가설을 기각합니다.")
else :
    print("귀무가설을 채택합니다.")
```
