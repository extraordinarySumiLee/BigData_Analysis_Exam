# 작업형 제1유형

다음 데이터에서 conventional 칼럼의 특수문자를 제거하고, IQR (3분위수 - 1분위수)를 구하시오. (소수점 첫째 자리에서 반올림)  

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/5_1_1.csv")
display(df.head(3))
```

```python
df['conventional'] = df['conventional'].str.replace(r'[^a-zA-Z0-9가-힣\s]',"",regex=True)
print(df['conventional'].unique())
```

```python
from sklearn.impute import SimpleImputer

df['conventional'] = df['conventional'].replace('',None)
imputer = SimpleImputer(strategy = 'most_frequent')
df[['conventional']] = imputer.fit_transform(df[['conventional']])
display(df.head(5))
print(df.info())
```

```python
df['conventional'] = pd.to_numeric(df['conventional'],errors='coerce')
Q1 = df['conventional'].quantile(0.25)
Q3 = df['conventional'].quantile(0.75)
print(f"IQR : {round(Q3-Q1,1)}")
```

위 데이터에서 흑인 또는 히스패닉 비율 (minority) / 빈곤율 (poverty) > 2 이며, 도시 유형 (city) 이 state 인 도시의 범죄율 (crime) 평균을 구하시오. (소수점 첫째자리에서 반올림)

```python
display(df.head(3))
df['minority ratio'] = round(df['minority']/df['poverty'],1)
df2 = df[(df['minority ratio']>2) & (df['city']=='state')]
display(f"소수 인종 및 도시유형 평균 : {round(df['crime'].mean(),1)}")
```

# 작업형 제2유형

훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오 (Target : 1일 확률)  

제출형식은 ID, prob 두 칼럼만 존재해야 한다. (평가지표  : AUC)

```python
import pandas as pd 
train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/5_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/5_2_test.csv")
```

```python
# EDA
display(train.head(3))
display(test.head(3))

print(train.info())
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
display(train['fractal_dimension_error'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/5_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/5_2_test.csv")

imputer = SimpleImputer(strategy='most_frequent')
train[['fractal_dimension_error']] = imputer.fit_transform(train[['fractal_dimension_error']])
test[['fractal_dimension_error']] = imputer.transform(test[['fractal_dimension_error']])

train_X = train.drop(['target','level_0','ID'],axis=1)
train_y = train['target']
test_X = test.drop(['target','level_0','ID'],axis=1)
test_y = test['target']

train_X, valid_X, train_y, valid_y = train_test_split(train_X,train_y,test_size=0.3,random_state=42)

from sklearn.ensemble import RandomForestClassifier # AUC 는 Classifier
rf = RandomForestClassifier()
rf.fit(train_X,train_y)

# 검증데이터로 미리 파악
from sklearn.metrics import roc_auc_score
valid_predict_proba = rf.predict_proba(valid_X)[:,1] # 정답이 1일 확률
valid_auc = roc_auc_score(valid_y,valid_predict_proba)
print(f"검증된 auc 점수 : {round(valid_auc,1)}")

# Test 검수
prediction = rf.predict_proba(test_X)[:,1]
AUC = roc_auc_score(test_y,prediction)
print(f"test 데이터의 AUC: {round(AUC,3)}")
test_id = test['ID']

result = pd.DataFrame({'ID' : test_id, 'prob' : prediction})
display(result.head(3))
```
