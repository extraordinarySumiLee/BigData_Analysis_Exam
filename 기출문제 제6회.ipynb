100 명의 키와 몸무게를 조사하여 적정 체중인지 판단할 수 있는 BMI 를 산출하려 한다. 아래 표를 참고하여 BMI를 기준으로 저체중, 정상, 과체중, 비만을 구분하고, 저체중인 사람과 비만인 사람의 총 합을 구하시오.
- 저체중 : BMI 18.5 미만
- 정상 : 18.5 이상 23 미만
- 과체중 : 23 이상 25 미만
- 비만 : 25 이상

$BMI = \frac{\text{몸무(kg)}}{\text{키}(m^2)}$ 
* 소숫점 첫째 자리까지 반올림

```python
import pandas as pd 

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_1_2.csv")
display(df.head(2))
```

```python
df['BMI'] = round(df['Weight_kg']/((df['Height_cm']*1E-2)**2),1)

def classification(BMI):
    if BMI < 18.5 :
        return "저체중"
    elif 18.5 <= BMI < 23 :
        return "정상"
    elif 23<= BMI < 25 :
        return "과체중"
    else :
        return "비만"

df['비만도결과'] = df['BMI'].apply(classification)

df2 = df[(df['비만도결과']=='저체중') | (df['비만도결과']=='비만')]
print(f"저체중인 사람과 비만인 사람의 총 합은 {len(df2)} 명 입니다.")
```

```python
# 정답 코드 
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_2_test.csv")

test_id = test['ID']
train = train.drop(['level_0','ID'],axis=1)
test = test.drop(['level_0','ID'],axis=1)

# 데이터 분할
train_X = train.drop(['DBP'],axis=1)
train_y = train['DBP']
test_X = test.drop(['DBP'],axis=1)
test_y = test['DBP']

from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X,train_y,test_size=0.3,random_state=42)

# Preprocessing
cat_cols = train_X.select_dtypes('object').columns.to_list()
num_cols = train_X.select_dtypes('number').columns.to_list()

# One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

onehotencoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
imputer = SimpleImputer(strategy = 'most_frequent')

train_X_categorical_imputed = imputer.fit_transform(train_X[cat_cols])
valid_X_categorical_imputed = imputer.transform(valid_X[cat_cols])
test_X_categorical_imputed = imputer.transform(test_X[cat_cols])

train_X_categorical_encoded = onehotencoder.fit_transform(train_X_categorical_imputed)
valid_X_categorical_encoded = onehotencoder.transform(valid_X_categorical_imputed)
test_X_categorial_encoded = onehotencoder.transform(test_X_categorical_imputed)

import numpy as np 
train_X_preprocessed = np.concatenate([train_X[num_cols], train_X_categorical_encoded],axis=1)
valid_X_preprocessed = np.concatenate([valid_X[num_cols], valid_X_categorical_encoded],axis=1)
test_X_preprocessed = np.concatenate([test_X[num_cols], test_X_categorial_encoded],axis=1)

# 모델 적합
from sklearn.ensemble import RandomForestRegressor 
rf = RandomForestRegressor(random_state=42)
rf.fit(train_X_preprocessed, train_y)
pred_val = rf.predict(valid_X_preprocessed)
# 모형 성능 확인
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error(valid_y, pred_val)
rmse = np.sqrt(mse)
print("valid RMSE : ",rmse)

# 테스트 데이터로 예측
test_pred = rf.predict(test_X_preprocessed)
test_pred = pd.DataFrame(test_pred, columns=['pred'])

result  = pd.concat([test_id,test_pred],axis=1)
print(result.head(2))
```

1. 직원들의 업무 수행 시간의 평균과 표준편차르 구하시오 (소수점 셋째 자리까지 반올림)  
2. 직원들의 업무 수행 시간이 정규분포를 따르는지 K-S 검정을 실시하고, 검정통계량을 계산하시오 (소수점 셋째 자리까지 반올림)  
3. p-value 를 바탕으로 유의수준 5% 에서 귀무가설의 기각/ 채택 여부를 결정하시오. (p-value 는 소수점 셋째 자리까지 반올림)  

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_3_1.csv")
display(df.head(3))
```

```python
print(f"직원들의 업무수행시간의 평균은 {round(df['work_hours'].mean(),3)} 이고 표준편차는 {round(df['work_hours'].std(),3)} 입니다.")
```

```python
from scipy.stats import kstest, norm 


statistic, pvalue = kstest(df['work_hours'],'norm',args = (df['work_hours'].mean(),df['work_hours'].std()))

print(f"K-S test 검정 통계량 : {round(statistic,3)}")
```

```python
alpha =0.05

if pvalue < 0.05 : 
    print("귀무가설을 기각합니다.")
else : 
    print("귀무가설을 채택합니다.")
```

다음의 데이터는 주택들의 가격(price), 면적(area), 방의 개수(rooms), 연식(age)을 조사하여 기록한 것이다.  
1. 주택 가격을 종속변수로 하고, 면적, 방의 개수, 연식을 독립변수로 하는 다중회귀 분석을 수행하여, 회귀계수가 가장 높은 변수를 구하시오. (다중회귀모형 적합 시 절편 포함)  
2. 유의수준 5% 하에서 각 독립변수가 주택 가격에 미치는 영향이 통계적으로 유의미한 지 판단하고, 유의미한 변수 개수를 구하시오.  

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_3_2.csv")
display(df.head(3))
```

```python
import statsmodels.api as sm 
# 독립 변수와 종속 변수 설정
X = df[['area','rooms','age']]
y = df['price']

# 상수항 추가
X = sm.add_constant(X)

# 다중회귀 분석 모델 적합
model = sm.OLS(y, X).fit()

# 회귀계수 추출
coefficients = model.params[1:]
print("회귀계수가 가장 큰 변수 : ",coefficients.idxmax())

p_values = model.pvalues
import numpy as np 
print("유의미한 변수 개수 : ",np.sum(p_values<0.05))
```

```python
import pandas as pd 
df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/6_3_2.csv")
import statsmodels.formula.api as smf

formula = 'price ~ area + rooms + age'
model = smf.ols(formula=formula,data=df).fit()
coefficients = model.params[1:]
print("회귀계수가 가장 큰 변수 : ",coefficients.idxmax())

pvalue = model.pvalues
import numpy as np 
print("유의미한 변수 개수 : ",np.sum(pvalue<0.05))
```
