3. 분기별 총 판매량 (제품 A~E의 합계) 의 월평균을 구하고, 월평균이 최대인 연도와 분기를 구하시오

```python
import pandas as pd 

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_1_3.csv")
display(df.head(3))
```

```python
df['총판매량'] = df['제품A']+df['제품B']+df['제품C']+df['제품D']+df['제품E']
display(df.head(3))
df['월'] = df['기간'].str.split('_').str[1]

def get_quarter(month) :
    if month in ['1월','2월','3월']:
        return '1분기'
    elif month in ['4월','5월','6월']:
        return '2분기'
    elif month in ['7월','8월','9월']:
        return '3분기'
    else :
        return '4분기'
df['분기'] = df['월'].apply(get_quarter)
display(df.head(5))
# 분기별 총 판매량의 월평균
display(df.groupby(['분기'])['총판매량'].mean())

# 월평균이 최대인 연도와 분기 구하기
#df2 = df.groupby(['기간'])['총판매량'].mean()
```

# 작업형 제2유형

훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오 (분류 예측값 제출)
** 제출 형식은 ID, pred 두 칼럼만 존재해야한다. (평가지표 : macro f1-score)

```python
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_2_test.csv")

display(train.head(3))
```

```python
# EDA
print("train data ")
print(train.isnull().sum())
print("test data ")
print(test.isnull().sum())
```

```python
display(train.head(3))
display(test.head(3))
```

```python
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_2_test.csv")

cat_cols = train.select_dtypes('object').columns.to_list()

if 'Target' in cat_cols:
    cat_cols.remove('Target')

for col in cat_cols:
    train[col] = train[col].astype(str).str.replace("_","",regex=False)
    test[col]= test[col].astype(str).str.replace("_","",regex=False)

train_X = train.drop(['Target'],axis=1)
train_X = train_X.astype(float)
test_X = test.drop(['Target'],axis=1)
test_X = test_X.astype(float)
train_y= train['Target']
test_y = test['Target']

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train_X,train_y)

pred =rf.predict(test_X)

from sklearn.metrics import f1_score
print("macro f1 score : ",f1_score(test_y,pred,average='macro'))

id_cols = test['ID']

df_result = pd.DataFrame({'ID' : id_cols, 'pred' : pred})
display(df_result)
```

# 작업형 제3유형

문제1 어느 학교에서 50명의 학생들을 대상으로 새로운 학습 프로그램의 효과를 조사하고자한다. 각 학생에 대해 학습 전과 학습 후의 시험 점수를 측정하였다.   


(1) 학습 전과 학습 후의 시험 점수의 평균과 표준편차를 구하시오 (소숫점 둘째 자리까지 반올림)  
(2) 학습 전후의 점수 차이가 유의미한지 검정하기 위해 대응표본 t-검정을 수행하고, 검정통계량을 계산하시오.  (소숫점 둘째 자리까지 반올림)  
(3) p-value를 바탕으로 유의수준 5%에서 귀무가설의 기각/채택 여부를 결정하시오. (p-value 는 소숫점 셋째 자리까지 반올림) 

```python
import pandas as pd 

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_3_1.csv")
display(df.head(3))
```

```python
# 학습 전과 학습 후의 시험점수의 평균과 표준편차
print(f"학습 전의 시험점수 평균은 {round(df['before'].mean(),2)} 이고 표준편차는 {round(df['before'].std(),2)} 입니다. ")
print(f"학습 후의 시험점수 평균은 {round(df['after'].mean(),2)} 이고 표준편차는 {round(df['after'].std(),2)} 입니다. ")

# 학습 점수 t-검정
from scipy import stats

st,pv = stats.ttest_rel(df['before'],df['after'])
print(f"학습 전후의 검정 통계량은 {round(st,2)} 입니다.")

# p_value 바탕
if pv < 0.05 : 
    print("귀무가설을 기각합니다.")
else :
    print("귀무가설을 채택합니다.")
```

문제2 어느 회사가 조사한 고객 데이터는 100개의 샘플로 이루어져 있으며, 각 샘플에는 고객의 나이, 소득, 가족 수, 그리고 제품 구매 여부가 포함되어 있다. 로지스틱 회귀 분석을 통해 고객의 제품 구매 여부를 예측하고자 한다. (임계값 0.5 기준)  

(1) 로지스틱 회귀 분석을 수행하고, 소득 변수의 오즈비를 계산하시오.  
(2) train 데이터 기준의 잔차 이탈도 (Residual Deviance) 를 계산하시오.  
(3) test 데이터로 오분류율을 계산하시오.  

```python
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_3_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/7_3_2_test.csv")
display(train.head(3))
```

```python
from statsmodels.formula.api import logit 
formula = "purchase ~ age + income + family_members"
model = logit(formula, data=train).fit()

# 오즈비 계산
import numpy as np 
odds_ratio = np.exp(model.params['income'])
print(f"소득 변수의 오즈비 : {round(odds_ratio,2)}")

# train 데이터의 잔차 이탈도 계산
print(f"잔차이탈도 : {round(-2*model.llf,2)}")
#print(model.deviance)
#print(f"Deviance 이용해 잔차이탈도 계산하면 : {round(model.deviance,2)}")
# test 데이터로 오분류율 계산 
pred = model.predict(test)
pred_class = (pred>=0.5).astype(int)

misclassification_rate = np.mean(test['purchase']!= pred_class)
print(f" Test 데이터 기준 오분류율 : {round(misclassification_rate, 4)}")
```
