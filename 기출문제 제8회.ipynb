# 작업형 제1유형

1. 다음의 데이터는 대륙별 국가의 맥주소비량을 조사한 것이다.   

(1) 평균 맥주소비량이 가장 많은 대륙을 구하시오  
(2) 이전 문제에서 구한 대륙에서 5번째로 맥주소비량이 많은 나라를 구하시오  
(3) 이전 문제에서 구한 나라의 평균 맥주소비량을 구하시오 (소수첨 첫째 자리에서 반올림)  

```python
import numpy as np 
import pandas as pd

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_1_1.csv")
display(df.head(3))
```

```python
df2 = df.groupby(['대륙'])['맥주소비량'].mean().reset_index()
display(df2)

# 1. 평균 맥주소비량이 가장 큰 행의 '정수 인덱스(방 번호)'를 찾습니다. (예: 3)
max_idx = df2['맥주소비량'].idxmax()

# 2. .loc[행_인덱스, '열_이름'] 구조로 대륙 이름을 쏙 빼옵니다.
print("평균 맥주소비량이 가장 많은 대륙 :")
print(df2.loc[max_idx, '대륙'])
# df.loc[행,열]
```

```python
# SA 대륙에서 5번째로 맥주소비량이 많은 국가
df3 = df[(df['대륙']=='SA')]
df4 = df3.groupby(['국가'])['맥주소비량'].sum().reset_index()
df4 = df4.sort_values(['맥주소비량'],ascending=False).reset_index()
display(df4)
print("SA 대륙에서 5번째로 맥주소비량이 많은 국가 : ")
print(df4.loc[4,"국가"])
```

```python
# Venezuela 의 평균 맥주 소비량 
df5 = df[(df['국가']=='Venezuela')]
print("Venezuela 의 평균 맥주 소비량 : ") 
ven_min = df5['맥주소비량'].mean()
print(round(ven_min,1))
```

2. 다음의 데이터는 국가별로 방문객 유형을 조사한 것이다.  

(1) 관광객비율이 두 번째로 높은 나라의 '관광' 수를 구하시오    
관광객 비율 = 관광 / 합계 (소수점 넷째 자리에서 반올림)  
합계 = 관광 + 사무 + 공무 + 유학 + 기타   

(2) 관광 수가 두번째로 높은 나라의 '공무'수의 평균을 구하시오 (소수점 첫째 자리에서 반올림)   

(3) 이전에 구한 관광 수와 공무 수의 합계를 구하시오  

```python
import numpy as np 
import pandas as pd

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_1_2.csv")
display(df.head(3))
```

```python
df['합계'] = df['관광'] + df['사무'] + df['공무'] + df['유학'] + df['기타']
df['관광객 비율'] = round(df['관광'] / df['합계'],4)
df2 = df.sort_values(['관광객 비율'],ascending=False).reset_index()

display(df2)
print("관광객 비율이 두 번째로 높은 나라의 '관광' 수 : ")
print(df2.iloc[1]['관광'])
```

```python
# 관광 수가 두번째로 높은 나라의 '공무' 수의 평균 
df3 = df[(df['국가']=='이탈리아')]
display(df3)
print("이탈리아 공무 수 평균 : ",round(df3['공무'].mean(),1))
```

```python
# 이전에 구한 관광 수와 공무 수의 합계 
print((df2.iloc[1]['관광'])+round(df3['공무'].mean(),1))
```

3. CO(GT),NMHC(GT) 칼럼에 대해서 Min-Max 스케일러를 실행하고, 스케일링된 CO(GT), NMHC(GT) 칼럼의 표준편차를 구하시오. (소숫점 셋째 자리에서 반올림 )

# 작업형 제2유형 

훈련 데이터로 학습한 모델을 테스트 데이터에 적용하여 예측한 결과를 제출하시오 (Target : count)   
제출 형식은 ID, pred 두 칼럼만 존재해야 한다. (평가지표 : MAE)  

```python
import numpy as np 
import pandas as pd

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_2_test.csv")
display(train.head(3))
display(test.head(3))
```

```python
# EDA 

print(train.info())
```

```python
# 최종 결과 제출을 위해서 ID 칼럼을 따로 저장합니다. 
test_id = test['ID']
```

```python
# 모델 성능 확인을 위해 훈련 데이터의 일부를 검증 데이터로 나눈다.
train_X = train.drop(['count'],axis=1)
train_y = train['count']

test_X = test.drop(['count'],axis=1)
test_y = test['count']

from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y= train_test_split(train_X, train_y, random_state=42, test_size=0.3)
print(train_X.shape, valid_X.shape, train_y.shape, valid_y.shape)
display(train_X.head(3))
display(valid_X.head(3))
```

```python
# 범주형 변수에 원핫 인코딩 수행
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

cat_columns = train_X.select_dtypes('object').columns.to_list()
print("cat columns : ",cat_columns)
train_X_pre = onehotencoder.fit_transform(train_X[cat_columns])
valid_X_pre = onehotencoder.fit_transform(valid_X[cat_columns])
test_X_pre = onehotencoder.fit_transform(test_X[cat_columns])
```

```python
# RandomForest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
rf.fit(train_X_pre,train_y)
```

```python
# 검증 데이터를 활용하여 모형 성능 확인
from sklearn.metrics import mean_absolute_error 

pred_val = rf.predict(valid_X_pre)
print("valid MAE : ",mean_absolute_error(valid_y,pred_val))
```

```python
# 테스트 데이터를 활용하여 최종 예측 
test_pred = rf.predict(test_X_pre)
test_pred = pd.DataFrame(test_pred, columns = ['pred'])
```

```python
# ID, pred 칼럼만 존재하는 결과 파일 생성
result = pd.concat([test_id, test_pred],axis=1)
print(result)
# result.to_csv('이름',index=False)
```

# 작업형 제3유형

1. 어느 회사에서 직원들의 업무 효율성을 높이기 위한 새로운 소프트웨어를 도입하였다. 도입 전과 도입 후의 업무 처리 시간을 각각 측정하여 새로운 소프트웨어의 효과를 검증하고자 한다.

(1) 도입 전과 도입후의 업무처리 시간의 평균과 표준편차를 구하시오. (소숫점 둘째 자리까지 반올림)  
(2) 도입 전후의 업무처리 시간 차이가 유의미한지 부호 순위 검정을 실시하고, 검정통계량을 계산하시오. (소숫점 둘째 자리까지 반올림)  
(3) p-value 를 바탕으로 유의수준 5% 에서 귀무가설의 기각/채택 여부를 결정하시오. (p-value 는 소숫점 셋째 자리까지 반올림)  


```python
import numpy as np 
import pandas as pd

df = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_3_1.csv")
display(df.head(3))
```

```python
# 도입 전과 후의 평균과 표준편차
print("도입 전의 평균 : {:.2f}, 표준편차 : {:.2f}".format(df['before'].mean(),df['before'].std()))
print("도입 후의 평균 : {:.2f}, 표준편차 : {:.2f}".format(df['after'].mean(),df['after'].std()))
```

```python
# 도입 전후 시간 차이 부호 순위 검정
from scipy.stats import wilcoxon

statistic, p_value = wilcoxon(df['before'],df['after'])
print("검정통계량 : {:.2f}".format(statistic))
```

```python
# p-value 바탕으로 귀무가설 기각/채택 여부
if p_value<0.05 : 
    print("귀무가설 기각")
else : 
    print("귀무가설 채택")
```

2. 어느 회사에서 직원들의 생산성에 영향을 미치는 요인이 무엇인지 확인하고자 한다. 100명의 직원들을 대상으로 생산성 점수, 근무 시간, 연령, 그리고 경력을 조사하였다.  

(1) 훈련 데이터를 기준으로 생산성 점수 (productivity) 를 종속변수로 하고 근무 시간, 연령, 그리고 경력을 독립변수로 하는 다중회귀 분석을 수행한 후, 회귀계수가 가장 높은 변수를 구하시오. (다중회쉬모형 적합 시 절편 포함)  
(2) 유의수준 5% 하에서 각 독립변수가 생산성에 미치는 영향이 통계적으로 유의미한지 판단하고, 유의미한 변수 개수를 구하시오 (p-value 는 소숫점 넷째자리까지 반올림)  
(3) 테스트 데이터로 모델의 성능을 평가하시오 (R^2 산출)

```python
import pandas as pd 

train = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_3_2_train.csv")
test = pd.read_csv(r"https://raw.githubusercontent.com/YoungjinBD/data/refs/heads/main/exam/8_3_2_test.csv")
display(train.head(3))
display(test.head(3))
```

```python
import statsmodels.api as sm

# 독립 변수와 종속 변수 설정 (train 데이터 사용) 
X_train = train[['hours','age','experience']]
y_train = train['productivity']

X_train = sm.add_constant(X_train) # 상수항 추가 
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
# 회귀계수가 가장 큰 변수를 출력 
coefficients = model.params[1:]
print("회귀계수가 가장 큰 변수 : ",coefficients.idxmax())
```

```python
# 유의수준 5% 하에서 p-value 를 확인하고, 유의미한 변수의 개수를 계산
p_values = model.pvalues[1:].round(4)
significant_vars = p_values[p_values<0.05]
num_significant_vars = len(significant_vars)

print("유의미한 변수 개수 : ",num_significant_vars)
print(significant_vars)
```

```python
# R^2 
# 독립 변수와 종속 변수 설정 (test 데이터 사용)
from sklearn.metrics import r2_score
X_test = test[['hours','age','experience']]
y_test = test['productivity']

X_test = sm.add_constant(X_test) # 상수항 추가
y_pred = model.predict(X_test) # 에측 수행
r2 = r2_score(y_test, y_pred) # R^2 계산
print("테스트 데이터 Rsquared : ",round(r2,3))
```
